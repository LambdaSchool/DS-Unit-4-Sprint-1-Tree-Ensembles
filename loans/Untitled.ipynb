```python

def clean(dat: pd.DataFrame) -> pd.DataFrame: 
    return (dat.dropna(axis=1)
           .assign(int_rate = dat.int_rate.apply(lambda s: float(s[:-1]))))

df = clean(pd.read_csv('train_features.csv'))

df.head()
```
