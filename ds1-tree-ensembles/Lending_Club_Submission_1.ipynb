```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

X_train = pd.read_csv('train_features.csv', index_col='id')
y_train = pd.read_csv('train_labels.csv', index_col='id')
X_test = pd.read_csv('test_features.csv', index_col='id')
    
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
pd.set_option('display.max_colwidth', -1)

X_train['training_set'] = 1
X_test['training_set'] = 0

X_full = pd.concat([X_train, X_test])

X_full.head()
```

```python
y_train.head()
```

```python
X_full_num = X_full.select_dtypes(include=['number']).copy()
X_full_num.head()
```

```python
X_full_num.tail()
```

```python
X_full_num.dtypes
```

```python
X_full_num = X_full_num.dropna(axis=1)
X_full_num.isna().sum()
```

```python
X_train = X_full_num[X_full_num['training_set'] == 1]
X_train = X_train.drop('training_set', axis=1)

X_test = X_full_num[X_full_num['training_set'] == 0]
X_test = X_test.drop('training_set', axis=1)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split

tree = DecisionTreeClassifier(criterion='gini',
                             max_depth=4, random_state=1)

tree.fit(X_train, y_train)

preds = tree.predict(X_test)

lending_club_sub = pd.DataFrame({'id': X_test.index, 'charged_off': preds})

lending_club_sub.to_csv('submission_1.csv', index=False)
```
